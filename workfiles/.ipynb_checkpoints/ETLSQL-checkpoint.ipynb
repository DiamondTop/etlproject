```python
ABBV = web.get_data_yahoo("ABBV",
                            start = "2009-01-01",
                            end = "2019-01-01")
```

```python
print(ABBV.head())
```

```python
a = pd.DataFrame(ABBV)
a.head()
```

```python
abbv_daily_returns = a['Adj Close'].pct_change()

abbv_monthly_returns = a['Adj Close'].resample('M').ffill().pct_change()
```

```python
abbv_df = pd.DataFrame(abbv_monthly_returns)
```

```python
abbv_df.head()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(abbv_df)
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("ABBV monthly returns data")
plt.show()
```
